```python
import pandas as pd

# Fixed input file for Notebook 3
file_path = "Processed_Reviews_with_Sentiment.csv"
df = pd.read_csv(file_path)

print("Loaded:", file_path)
print("Shape:", df.shape)

# Find sentiment column robustly (case-insensitive)
sentiment_col = next((c for c in df.columns if c.strip().lower() == "sentiment"), None)
if sentiment_col is None:
    raise ValueError(
        "Could not find a 'Sentiment' column. Available columns: " + ", ".join(df.columns)
    )

# Normalize values to POSITIVE / NEGATIVE / NEUTRAL
label_map = {"LABEL_0": "NEGATIVE", "LABEL_1": "NEUTRAL", "LABEL_2": "POSITIVE"}
normalized = (
    df[sentiment_col]
    .astype(str)
    .str.strip()
    .str.upper()
    .replace(label_map)
)

# Keep expected classes + missing/other checks
order = ["POSITIVE", "NEGATIVE", "NEUTRAL"]
counts = normalized.value_counts(dropna=False)
dist = counts.reindex(order, fill_value=0)

total = len(df)
result = pd.DataFrame({
    "Count": dist,
    "Percent": (dist / total * 100).round(2),
})

print("\nP/N/NE Sentiment Distribution:")
print(result)

other_count = int((~normalized.isin(order)).sum())
if other_count > 0:
    print(f"\nOther/Unexpected labels: {other_count}")

missing_count = int(df[sentiment_col].isna().sum())
if missing_count > 0:
    print(f"Missing sentiment values: {missing_count}")
```
